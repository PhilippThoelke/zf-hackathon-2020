```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../datasets/ts3_3_k_3.0.csv', names=['time', 'distance', 'height'], header=None, index_col=0)
df = df.loc[:80]
```

```python
df['distance'].plot(title='distance', figsize=(15, 5))
plt.show()
df['height'].plot(title='height', figsize=(15, 5))
plt.show()
```

```python
mean = df.rolling(100).mean()[1::100]
((mean['distance'] - mean['distance'].shift(1)) / 0.005).plot(title='velocity', figsize=(15, 5))
plt.show()
```
